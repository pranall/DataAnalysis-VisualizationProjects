# ***Key Performance Indicators (KPIs)***

To effectively measure the success of the public awareness campaign on particulate matter based on the data visualized in thedashboard, here are several **Key Performance Indicators (KPIs)** that could be tracked:

### 1. **Air Quality Improvement Indicators**
   - **Reduction in PM2.5 and PM10 levels**: Track the change in average PM2.5 and PM10 levels before and after the campaign, particularly during high-risk seasons (winter and spring). A reduction in these levels would indicate that the campaign has helped raise awareness about the sources of particulate pollution and encouraged practices to reduce emissions or exposure.
     - **Formula**: `(Post-campaign PM2.5 or PM10 Level - Pre-campaign PM2.5 or PM10 Level) / Pre-campaign PM2.5 or PM10 Level * 100`
     - **Target**: Achieve a X% reduction in PM2.5/PM10 levels during the campaign period.

### 2. **Awareness and Engagement Indicators**
   - **Increase in Public Awareness**: Measure the increase in public knowledge about air quality, particulate matter, and health risks. This could be measured through surveys or social media analytics (mentions, shares, and hashtag usage related to the campaign).
     - **Survey Results**: Pre- and post-campaign surveys that assess public understanding of PM2.5/PM10 and health risks.
     - **Target**: Achieve a Y% increase in the percentage of people who are aware of the health risks posed by particulate matter.

   - **Engagement with Campaign Materials**: Measure the number of views, clicks, shares, and interactions with digital content (videos, articles, infographics) created as part of the campaign. This can include social media posts, website visits, and campaign video views.
     - **Formula**: `Total Interactions (views, clicks, shares) / Total Impressions * 100`
     - **Target**: Achieve a Z% increase in engagement compared to a baseline or previous campaigns.

### 3. **Behavioral Change Indicators**
   - **Increase in Use of Air Filtration Systems**: Track the increase in sales or installations of air filtration devices or air purifiers, especially during the winter months when PM2.5 levels tend to be higher.
     - **Formula**: `(Post-campaign installations - Pre-campaign installations) / Pre-campaign installations * 100`
     - **Target**: Achieve a W% increase in sales of air filtration systems or air quality monitoring devices.

   - **Decrease in Outdoor Activities on High Pollution Days**: Measure the decrease in outdoor activities (such as physical activities, travel, or public gatherings) on days with high air pollution.
     - **Formula**: `(Number of outdoor activities on high pollution days pre-campaign - post-campaign) / Pre-campaign number of outdoor activities`
     - **Target**: Achieve a V% reduction in outdoor activities on high-pollution days.

### 4. **Health Impact Indicators**
   - **Reduction in Respiratory Health Issues**: Track changes in the number of hospital visits, emergency room visits, or doctor visits related to respiratory issues, such as asthma, COPD, and other particulate matter-related conditions. This would require collaboration with healthcare providers to track these metrics.
     - **Formula**: `(Pre-campaign respiratory issues - Post-campaign respiratory issues) / Pre-campaign respiratory issues * 100`
     - **Target**: Achieve a decrease in respiratory health issues linked to particulate matter by a certain percentage.

   - **Increased Usage of Health Resources**: Measure how many people use resources provided by the campaign, such as guidelines for protecting against particulate matter or the use of personal protective equipment (e.g., masks).
     - **Formula**: `Number of people using health resources / Total target audience * 100`
     - **Target**: Achieve a W% usage rate of campaign-provided health resources.

### 5. **Seasonal Impact Indicators**
   - **Change in Seasonal Air Quality Perception**: Track changes in the publics perception of air quality during different seasons through surveys or sentiment analysis.
     - **Formula**: `Post-campaign perception of air quality - Pre-campaign perception of air quality`
     - **Target**: Achieve a positive shift in public perception of air quality during high-pollution seasons (winter and spring).

   - **Improved Seasonal Health Metrics**: Track changes in health outcomes related to air quality by season, particularly in winter and spring when particulate pollution is highest.
     - **Formula**: `Improvement in health outcomes during high-pollution seasons`
     - **Target**: Achieve improved health outcomes (fewer asthma attacks, hospital admissions) in the target population during high-risk seasons.

### 6. **Public Engagement and Advocacy Indicators**
   - **Campaign Reach and Coverage**: Measure the geographical and demographic reach of the campaign, ensuring that areas with the most particulate pollution are effectively targeted.
     - **Formula**: `Total audience reached / Targeted population * 100`
     - **Target**: Achieve broad reach, particularly in high-risk areas or demographics.

   - **Advocacy and Community Action**: Measure the number of community-driven actions or advocacy groups formed as a result of the campaign, such as local organizations or schools implementing their air quality programs or encouraging others to reduce exposure.
     - **Formula**: `Number of new advocacy groups or actions / Total population * 100`
     - **Target**: Achieve a significant number of local community actions or programs focusing on reducing exposure to particulate matter.

By monitoring these KPIs, the Arizona Department of Health and Safety will be able to gauge the effectiveness of their public awareness campaign in terms of improving public awareness, changing behaviors, reducing exposure to particulate matter, and ultimately, improving health outcomes.

# ***Data Preparation for the dashboards***



***Step 1: Import all the necessary libraries***

```python
import pandas as pd
import seaborn as sns
from matplotlib import rcParams
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
from matplotlib.colors import LinearSegmentedColormap
from matplotlib.offsetbox import OffsetImage, AnnotationBbox
```

***Step 2: Load the Dataset***

```python
#Load dataset
data = pd.read_csv('Air Quality of Arizona.csv')
print("Columns in the dataset:", data.columns)
data.head()
```

***Step 3: Preprocess the Data***

Parse dates and extract useful features like month and season.
Identify good and poor air quality thresholds based on PM2.5 and PM10.

```python
# Rename columns for easier access (optional, just for clarity)
data.rename(columns={
    'Daily.Mean.PM2.5.Concentration.ug.m3': 'PM2.5',
    'Daily.PM2.5.AQI.Value': 'PM2.5_AQI',
    'Daily.Mean.PM10.Concentration.ug.m3': 'PM10',
    'Daily.PM10.AQI.Value': 'PM10_AQI',
    'SITE_LATITUDE': 'Latitude',
    'SITE_LONGITUDE': 'Longitude',
    'STATE': 'State'
}, inplace=True)

# Drop unnecessary columns (if not needed for analysis)
data.drop(columns=['Unnamed: 0', 'Site.ID'], inplace=True)

# Convert 'Date' column to datetime
data['Date'] = pd.to_datetime(data['Date'])

# Extract useful features: Month, Year, and Season
data['Month'] = data['Date'].dt.month
data['Year'] = data['Date'].dt.year

# Define a function to assign seasons based on months
def get_season(month):
    if month in [12, 1, 2]:
        return 'Winter'
    elif month in [3, 4, 5]:
        return 'Spring'
    elif month in [6, 7, 8]:
        return 'Summer'
    else:
        return 'Fall'

# Apply the function to assign seasons
data['Season'] = data['Month'].apply(get_season)

# Define air quality thresholds for classification
good_pm25_threshold = 12  # g/m, based on EPA standards
good_pm10_threshold = 50  # g/m, based on EPA standards
poor_pm25_threshold = 35  # g/m, upper limit for acceptable PM2.5
poor_pm10_threshold = 150 # g/m, upper limit for acceptable PM10

# Categorize air quality into 'Good', 'Moderate', or 'Poor'
def classify_air_quality(row):
    if row['PM2.5'] <= good_pm25_threshold and row['PM10'] <= good_pm10_threshold:
        return 'Good'
    elif row['PM2.5'] > poor_pm25_threshold or row['PM10'] > poor_pm10_threshold:
        return 'Poor'
    else:
        return 'Moderate'

data['AirQuality'] = data.apply(classify_air_quality, axis=1)

# Display the first few rows of the updated DataFrame for verification
print(data.head())
```

***Step 4: Aggregate Data for Visualization***

Group data by month and season to analyze trends.

```python
# Rename columns for easier access (if not already done in Step 3)
data.rename(columns={
    'Daily.Mean.PM2.5.Concentration.ug.m3': 'PM2.5',
    'Daily.Mean.PM10.Concentration.ug.m3': 'PM10',
    'Daily.PM2.5.AQI.Value': 'PM2.5_AQI',
    'Daily.PM10.AQI.Value': 'PM10_AQI'
}, inplace=True)

# Group data by season and calculate normalized air quality counts (percentage of days per category)
season_summary = (
    data.groupby('Season')['AirQuality']
    .value_counts(normalize=True)  # Normalize to get percentages
    .unstack()                     # Convert into a table format (rows: seasons, columns: air quality categories)
    .fillna(0)                     # Fill missing values with 0
)

# Display the season_summary to check calculations
print("Seasonal Air Quality Summary:")
print(season_summary)

# Group data by month and calculate mean PM2.5 and PM10 concentrations
monthly_avg = (
    data.groupby('Month')[['PM2.5', 'PM10']]
    .mean()              # Calculate mean concentrations
    .reset_index()       # Reset index to keep 'Month' as a column
)

# Display the monthly averages to check calculations
print("\nMonthly Average PM2.5 and PM10 Concentrations:")
print(monthly_avg)
```
