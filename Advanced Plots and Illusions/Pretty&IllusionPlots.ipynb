# ***Data preparation and code for all the plots***

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

#Load dataset
data = pd.read_csv('Slugs&Snails.csv')
print("Columns in the dataset:", data.columns)
data.head()
```

```python
#Example mapping of dates to months
spring_months = ['03', '04', '05']  # March, April, May
fall_months = ['09', '10', '11']   # September, October, November

#Extract season from the 'Date' column
def determine_season(date):
    month = date.split('-')[1]  # Extract the month as a string
    if month in spring_months:
        return 'Spring'
    elif month in fall_months:
        return 'Fall'
    else:
        return 'Other'

#Add the 'Season' column
data['Season'] = data['Date'].apply(determine_season)

#Sum the counts of mollusks by year (Time) and season
seasonal_counts = data.groupby(['Time', 'Season'])['Count'].sum().reset_index()

#Count the unique species by year and season
seasonal_species = data.groupby(['Time', 'Season'])['ScientificName'].nunique().reset_index()

#Rename the column for clarity
seasonal_species.rename(columns={'ScientificName': 'Species'}, inplace=True)

#Display processed data (optional for verification)
print(seasonal_counts.head())
print(seasonal_species.head())
```

```python
#Simulate loading and preprocessing the slugs and snails dataset
#Replace this with actual data loading and processing
data = {
    'Year': [2015, 2016, 2017, 2018, 2019, 2020, 2021] * 2,
    'Season': ['Spring'] * 7 + ['Fall'] * 7,
    'Count': [150, 200, 170, 180, 210, 190, 220, 140, 180, 160, 150, 190, 170, 200],
    'Species': [10, 12, 11, 11, 13, 12, 14, 8, 10, 9, 9, 11, 10, 12],
}
df = pd.DataFrame(data)

#Separate data for counts and species
counts = df[['Year', 'Season', 'Count']]
species = df[['Year', 'Season', 'Species']]

print(counts)
print(species)
```

```python
#Code for Pretty Plot
#Custom color palette (using green, teal, yellow, and brown colors)
custom_palette = sns.color_palette(["#2ca02c", "#17becf", "#ffcc00", "#8b4513"])  # Green, teal, yellow, brown

#Pretty Plot: Line plot for the counts of slugs and snails and species richness
fig, ax1 = plt.subplots(figsize=(12, 6))

#Line plot for the counts of slugs and snails
sns.lineplot(
    data=counts,
    x='Year',
    y='Count',
    hue='Season',
    marker='o',
    ax=ax1,
    linewidth=2,
    palette=custom_palette[:2],  #Green and Teal for the counts
)

#Adding second y-axis for species richness
ax2 = ax1.twinx()
sns.lineplot(
    data=species,
    x='Year',
    y='Species',
    hue='Season',
    marker='s',
    ax=ax2,
    linestyle='--',
    linewidth=2,
    palette=custom_palette[2:],  #Yellow and Brown for the species
)

#Titles and labels
ax1.set_title("Changes in Slug and Snail Counts and Species Richness Over Time", fontsize=16)
ax1.set_xlabel("Year", fontsize=14)
ax1.set_ylabel("Number of Slugs and Snails", fontsize=14)
ax2.set_ylabel("Number of Species Found", fontsize=14)

#Legends
handles, labels = ax1.get_legend_handles_labels()
ax1.legend(handles=handles, labels=labels, title="Season", loc="upper left")

#Grid lines for readability
ax1.grid(axis='y', linestyle='--', alpha=0.7)

#Layout adjustments to avoid overlaps and improve readability
plt.tight_layout()

#Save the pretty plot as high-resolution image (300 DPI for publication quality)
plt.savefig("pretty_plot_high_quality.png", dpi=300, bbox_inches="tight")

#Show the plot
plt.show()

#Caption for the plot
print("Figure 1: This plot shows the changes in the number of slugs and snails and the number of slug and snail species found by biologists over time. The data was collected from surveys conducted in the spring and fall between 2015 and 2021. The plot displays two distinct trends: Number of Slugs and Snails (Primary Y-axis): This is represented by the solid line with circular markers. The counts are generally higher in the spring, with a noticeable peak in 2019 before dropping again in 2021. Number of Species Found (Secondary Y-axis): This is represented by the dashed line with square markers, plotted on a secondary y-axis to avoid overlap with the counts. Species richness fluctuates similarly to the slug and snail counts, with a clear peak in 2021 in the fall.")
```

```python
#Code for Illusion Plot
#Define a distorted color palette (e.g., a reversed and conflicting ramp)
illusion_palette = sns.color_palette(["#8b4513", "#ffcc00", "#17becf", "#2ca02c"])  #Reverse order: brown, yellow, teal, green

#Manually apply square-root transformation for the secondary axis
species['Transformed_Species'] = species['Species'] ** 0.5  #Square-root transformation

#Plotting
fig, ax1 = plt.subplots(figsize=(12, 6))

#Line plot for the counts of slugs and snails with a distorted y-axis
sns.lineplot(
    data=counts,
    x='Year',
    y='Count',
    hue='Season',
    marker='o',
    ax=ax1,
    linewidth=2,
    palette=illusion_palette[:2],  #Misleading ramp (reversed from the pretty plot)
)
ax1.set_yscale("log")  #Apply a log scale to distort perception of differences

#Adding second y-axis for species richness (square-root transformed)
ax2 = ax1.twinx()
sns.lineplot(
    data=species,
    x='Year',
    y='Transformed_Species',
    hue='Season',
    marker='s',
    ax=ax2,
    linestyle='--',
    linewidth=2,
    palette=illusion_palette[2:],  #Use reversed colors here too
)
ax2.set_ylabel("Number of Species Found (Square-Root Transformed)", fontsize=14, labelpad=10)

#Titles and labels
ax1.set_title("Slug and Snail Counts and Species Richness Over Time (Illusion Plot)", fontsize=16, pad=15)
ax1.set_xlabel("Year", fontsize=14, labelpad=10)
ax1.set_ylabel("Number of Slugs and Snails (Log Scale)", fontsize=14, labelpad=10)

#Misleading legend
handles, labels = ax1.get_legend_handles_labels()
ax1.legend(handles=handles, labels=labels, title="Season", loc="lower center", fontsize=10, frameon=True)  #Awkward placement

#Adjust layout
plt.tight_layout(rect=[0, 0.2, 1, 1])  #Leave space for the caption

#Save the illusion plot
plt.savefig("illusion_plot_corrected.png", dpi=300, bbox_inches="tight")
plt.show()

#print the caption
print("Figure 2: This illusion plot shows the changes in the number of slugs and snails and the number of slug and snail species found over time. A reversed color ramp is used, and axis scales have been altered to logarithmic (slugs and snails) and square-root transformed (species richness), which may distort the interpretation of the trends. Differences between spring and fall seasons now appear exaggerated or diminished depending on the year and metric.")
```
